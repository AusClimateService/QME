```python
#bias adjusted
import time
import glob
import matplotlib.pyplot as plt
import xarray as xr

org="BOM" #CSIRO BOM
rcm="BARPA-R" #CCAM-v2203-SN BARPA-R
grid="AUST-20i"
# Start timing
start_time = time.time()
yr_start="1980-01-01"
yr_end="2014-12-31"
# Path to your directory containing NetCDF files
directory20= f'/g/data/xv83/users/at2708/bias_adjustment_acs_qme/qme_dev_hourly/outputs_netcdf2netcdf_contiguous_wbgt20/{org}/ACCESS-CM2/historical/r4i1p1f1/{rcm}/v1-r1-ACS-QME-BARRAR2-1980-2022/1hr/wbgtAdjust/v20250311/'
# Find all NetCDF files in the directory
file_list20 = glob.glob(directory20 + '*.nc')

# Open multiple NetCDF files as a single xarray Dataset
ds_adjust20 = xr.open_mfdataset(file_list20, combine='by_coords', engine='h5netcdf', chunks={})

ds_adjust_netcdf20 = ds_adjust20.wbgtAdjust.sel(time=slice(yr_start,yr_end))
ds_adjust_netcdf_mean20 = ds_adjust_netcdf20.mean("time").compute()

print(f"Processing took {time.time() - start_time:.2f} seconds")
```

```python
#historical

start_time = time.time()

# Path to your directory containing NetCDF files
directory20 = f'/g/data/ia39/australian-climate-service/test-data/CORDEX/output-CMIP6/DD/{grid}/{org}/ACCESS-CM2/historical/r4i1p1f1/{rcm}/v1-r1/1hr/wbgt/latest-conservative-contiguous/'

# Find all NetCDF files in the directory
file_list20 = glob.glob(directory20 + '*.nc')
# Open multiple NetCDF files as a single xarray Dataset
ds_hist20 = xr.open_mfdataset(file_list20, combine='by_coords', engine='h5netcdf', chunks={})

ds_hist_netcdf20 = ds_hist20.wbgt.sel(time=slice(yr_start,yr_end))
ds_hist_netcdf_mean20 = ds_hist_netcdf20.mean("time").compute()

print(f"Processing took {time.time() - start_time:.2f} seconds")
```

```python
#reference
start_time = time.time()

# Path to your directory containing NetCDF files
directory20 = f'/g/data/ia39/australian-climate-service/test-data/BARRA2/output/reanalysis/{grid}/BOM/ERA5/historical/hres/BARRAR2/v1/1hr/wbgt/latest-conservative-contiguous/'

# Find all NetCDF files in the directory
file_list20 = glob.glob(directory20 + '*.nc')

# Open multiple NetCDF files as a single xarray Dataset
ds_ref20 = xr.open_mfdataset(file_list20, combine='by_coords', engine='h5netcdf', chunks={})

ds_ref_netcdf20 = ds_ref20.wbgt.sel(time=slice(yr_start,yr_end))
ds_ref_netcdf_mean20 = ds_ref_netcdf20.mean("time").compute()

print(f"Processing took {time.time() - start_time:.2f} seconds")
```

```python
da1 = ds_ref_netcdf_mean20
da2 = ds_hist_netcdf_mean20 - ds_ref_netcdf_mean20
da3 = ds_adjust_netcdf_mean20 - ds_ref_netcdf_mean20
vmin = -1
vmax = 1
fig, axes = plt.subplots(1, 3, figsize=(18, 6), sharex=True, sharey=True)

da1.plot(
    ax=axes[0],
    vmin=5,
    vmax=27,
    cmap="hot_r",
    add_colorbar=True,
    cbar_kwargs={"label": "WBGT (C)"}
)
axes[0].set_title("RARRAR2 annual mean wbgt 1980-2014")
axes[0].set_xlabel("Longitude")
axes[0].set_ylabel("Latitude")

da2.plot(
    ax=axes[1],
    vmin=vmin,
    vmax=vmax,
    cmap="RdBu_r",
    add_colorbar=True,
    cbar_kwargs={"label": "Bias (C)"}
)
axes[1].set_title(f"Bias (model minus ref) in the 1980-2014 annual mean wbgt \n (raw ACCESS-CM2-{rcm})")
axes[1].set_xlabel("Longitude")
axes[1].set_ylabel("Latitude")

da3.plot(
    ax=axes[2],
    vmin=vmin,
    vmax=vmax,
    cmap="RdBu_r",
    add_colorbar=True,
    cbar_kwargs={"label": "Bias (C)"}
)
axes[2].set_title(f"Bias in mean wbgt 1980-2014 rel. to reference \n (bias-adjusted ACCESS-CM2-{rcm})")
axes[2].set_xlabel("Longitude")
axes[2].set_ylabel("Latitude")

plt.tight_layout()
plt.show()
fig.savefig(f"/g/data/xv83/users/at2708/bias_adjustment_acs_qme/qme_dev_hourly/outputs_netcdf2netcdf_contiguous_wbgt20/wbgt_{org}_{rcm}_bias-correction_{grid}_limitstasmax.png", dpi=300, bbox_inches='tight') 
```
